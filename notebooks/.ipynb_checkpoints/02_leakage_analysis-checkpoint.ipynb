<h2 style="font-size:32px; font-family:Georgia; font-weight:bold;">
Leakage Analysis: Patient‑Level Data Leakage
</h2>

<h3 style="font-size:24px; font-family:Georgia; font-weight:bold;">
Objective
</h3>

<p style="font-size:15px; font-family:Arial;">
This notebook investigates potential <b>data leakage risks</b> in the Chest X‑Ray dataset obtained from ActiveLoop. The analysis focuses on determining whether images belonging to the same patient (<code>person_num</code>) appear across both training and validation/test splits when a naïve image‑level split is applied.
</p>

<p style="font-size:15px; font-family:Arial;">
This evaluation is essential because:
</p>

<ul style="font-size:15px; font-family:Arial;">
  <li>Medical imaging datasets frequently contain <b>multiple images per patient</b></li>
  <li>Random image‑level splitting can inadvertently leak patient‑specific features</li>
  <li>Such leakage can produce <b>artificially inflated performance metrics</b></li>
</ul>

<p style="font-size:15px; font-family:Arial;">
The purpose of this notebook is <b>not</b> to train a model, but to <b>demonstrate whether leakage exists</b> and to <b>quantify its impact</b> on dataset integrity.
</p>

<hr>

<h3 style="font-size:24px; font-family:Georgia; font-weight:bold;">
Scope of Analysis
</h3>

<p style="font-size:15px; font-family:Arial;">
The following tasks are performed to assess leakage risk:
</p>

<ul style="font-size:15px; font-family:Arial;">
  <li>Loading the dataset from ActiveLoop</li>
  <li>Inspecting dataset tensors and metadata</li>
  <li>Extracting and analyzing patient identifiers (<code>person_num</code>)</li>
  <li>Measuring the number of images associated with each patient</li>
  <li>Demonstrating how naïve image‑level splits can cause leakage</li>
  <li>Documenting why such leakage invalidates model evaluation</li>
</ul>

<hr>

<h3 style="font-size:24px; font-family:Georgia; font-weight:bold;">
Libraries Used
</h3>

<ul style="font-size:15px; font-family:Arial;">
  <li><b>deeplake</b> – dataset loading and interaction</li>
  <li><b>numpy</b> – numerical operations and aggregation</li>
  <li><b>pandas</b> – tabular inspection and analysis</li>
  <li><b>matplotlib</b> – visualizations</li>
  <li><b>collections</b> – counting and distribution analysis</li>
</ul>

<hr>

```python
import deeplake
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import warnings
from collections import Counter
from sklearn.exceptions import ConvergenceWarning
warnings.filterwarnings("ignore", category=ConvergenceWarning)

plt.style.use("seaborn-v0_8")
```

<h2 style="font-size:28px; font-family:Georgia; font-weight:bold;">
Initial Tensor Inspection
</h2>

<p style="font-size:15px; font-family:Arial;">
The dataset was loaded from Deep Lake and inspected to verify structure and label integrity. Available tensors were listed, and sample count was confirmed.
</p>

<ul style="font-size:15px; font-family:Arial;">
  <li><b>5216 total samples</b> detected in the dataset</li>
  <li><b>Label tensor shape:</b> <code>(5216, 1)</code>, indicating one label per image</li>
  <li><b>Unique label values:</b> <code>[0, 1, 2]</code></li>
</ul>

<p style="font-size:15px; font-family:Arial;">
These findings confirm that the dataset is structured for image classification and contains three distinct diagnostic classes. No group metadata was present, and all tensors were accessed directly.
</p>


```python
# Load dataset
ds = deeplake.load(
    "hub://activeloop/chest-xray-train",
    read_only=True
)

ds
```

```python
print("Available tensors:")
ds.tensors.keys()

print("\nGroups:")
ds.groups
```

```python
num_samples = len(ds)
print(f"Total samples: {num_samples}")
```

```python
labels = ds.labels

labels.shape
```

```python
labels_list = labels.numpy(aslist=True)

# Flatten nested lists
labels_flat = [int(l[0]) if isinstance(l, (list, np.ndarray)) else int(l)
               for l in labels_list]

print("Unique label values:", sorted(set(labels_flat)))
```

<h2 style="font-size:28px; font-family:Georgia; font-weight:bold;">
Class Distribution Analysis
</h2>

<p style="font-size:15px; font-family:Arial;">
The label distribution was computed to assess class balance across the dataset. A frequency count was performed using <code>Counter</code>, followed by a bar chart visualization to highlight disparities between diagnostic categories.
</p>

<ul style="font-size:15px; font-family:Arial;">
  <li><b>Class 0 (Normal):</b> 1341 images</li>
  <li><b>Class 1 (Bacterial Pneumonia):</b> 2530 images</li>
  <li><b>Class 2 (Viral Pneumonia):</b> 1345 images</li>
</ul>

<p style="font-size:15px; font-family:Arial;">
The chart confirms that pneumonia cases dominate the dataset, with bacterial pneumonia being the most represented class. Normal and viral pneumonia classes are nearly balanced, but underrepresented relative to bacterial cases.
</p>

<p style="font-size:15px; font-family:Arial;">
This imbalance may influence model behavior and should be considered when designing loss functions, sampling strategies, or evaluation metrics.
</p>


```python
label_counts = Counter(labels_flat)

print("Class distribution:")
for label, count in label_counts.items():
    print(f"Class {label}: {count}")
```

```python
plt.figure(figsize=(5,4))
plt.bar(label_counts.keys(), label_counts.values())
plt.xticks([0, 1], ["Normal", "Pneumonia"])
plt.ylabel("Number of Images")
plt.title("Class Distribution")
plt.show()
```

<h2 style="font-size:28px; font-family:Georgia; font-weight:bold;">
Class-Specific Sample Visualization
</h2>

<p style="font-size:15px; font-family:Arial;">
Representative chest X-ray images were displayed for both the Normal and Pneumonia classes to visually assess anatomical differences and confirm label alignment. Three samples were shown per class using a filtered subset of the dataset.
</p>

<h3 style="font-size:20px; font-family:Georgia; font-weight:bold;">
Observations
</h3>

<ul style="font-size:15px; font-family:Arial;">
  <li><b>Normal samples</b> exhibit clear lung fields with no visible abnormalities</li>
  <li><b>Pneumonia samples</b> show signs of lung congestion, opacities, and reduced clarity in pulmonary regions</li>
  <li><b>Intra-class variability</b> is present, especially within pneumonia cases, reinforcing the need for robust feature extraction</li>
</ul>

<p style="font-size:15px; font-family:Arial;">
This visual inspection supports the validity of the label annotations and highlights the complexity of distinguishing pathological cases from normal anatomy.
</p>

```python
def show_samples(ds, label_value, title, n=3):
    indices = [i for i, l in enumerate(labels_flat) if l == label_value][:n]
    
    plt.figure(figsize=(10,3))
    for i, idx in enumerate(indices):
        plt.subplot(1, n, i+1)
        plt.imshow(ds.images[idx].numpy())
        plt.axis("off")
    plt.suptitle(title)
    plt.show()

show_samples(ds, label_value=0, title="Normal Chest X-Rays")
show_samples(ds, label_value=1, title="Pneumonia Chest X-Rays")
```

<h2 style="font-size:28px; font-family:Georgia; font-weight:bold;">
Patient Image Distribution
</h2>

<p style="font-size:15px; font-family:Arial;">
Patient identifiers were extracted and flattened to account for dynamic entries. A total of <b>5216 images</b> were associated with <b>1636 unique patients</b>, confirming that multiple samples exist per patient.
</p>

<p style="font-size:15px; font-family:Arial;">
A histogram was plotted to visualize the number of images per patient. The distribution reveals that most patients contribute only a few images, while a small subset have significantly more.
</p>

<h3 style="font-size:20px; font-family:Georgia; font-weight:bold;">
Implications
</h3>

<ul style="font-size:15px; font-family:Arial;">
  <li><b>Patient-level imbalance</b> must be addressed during split construction</li>
  <li><b>High-volume patients</b> pose a risk of leakage if not isolated</li>
  <li><b>Evaluation metrics</b> may be skewed if patient overlap occurs across splits</li>
</ul>

<p style="font-size:15px; font-family:Arial;">
This analysis reinforces the need for patient-level partitioning and motivates further inspection of split integrity.
</p>


```python
person_ids_raw = ds.person_num.numpy(aslist=True)

# Safely flatten dynamic entries
person_ids = []
for pid in person_ids_raw:
    if isinstance(pid, (list, np.ndarray)) and len(pid) > 0:
        person_ids.append(int(pid[0]))
    elif isinstance(pid, (int, np.integer)):
        person_ids.append(int(pid))
    else:
        person_ids.append(None)

print("Total images:", len(person_ids))
print("Unique patients:", len(set(person_ids)))
```

```python
patient_counts = Counter(person_ids)

counts = list(patient_counts.values())

plt.figure(figsize=(6,4))
plt.hist(counts, bins=30)
plt.xlabel("Images per Patient")
plt.ylabel("Number of Patients")
plt.title("Distribution of Images per Patient")
plt.show()
```

<h2 style="font-size:28px; font-family:Georgia; font-weight:bold;">
Patient Overlap and Leakage Verification
</h2>

<p style="font-size:15px; font-family:Arial;">
An analysis was performed to quantify how many patients contribute multiple images and to assess the risk of leakage under naïve splitting. Out of <b>1636 unique patients</b>, <b>1006 patients</b> were found to have more than one image, representing <b>61.49%</b> of the dataset.
</p>

<p style="font-size:15px; font-family:Arial;">
A random image-level split was then applied to simulate a naïve partitioning strategy. Patient identifiers were extracted from both training and test sets, and overlap was measured.
</p>

<ul style="font-size:15px; font-family:Arial;">
  <li><b>447 patients</b> were found to appear in both training and test sets</li>
  <li>This confirms the presence of <b>patient-level leakage</b> under image-level splitting</li>
</ul>

<p style="font-size:15px; font-family:Arial;">
These findings demonstrate that random image-based partitioning introduces significant leakage, invalidating any downstream model evaluation. Patient-level splitting is required to ensure disjoint partitions and reliable performance metrics.
</p>


```python
multi_image_patients = sum(1 for c in patient_counts.values() if c > 1)

print(f"Patients with multiple images: {multi_image_patients}")
print(f"Percentage of patients with multiple images: {multi_image_patients / len(patient_counts):.2%}")
```

```python
from sklearn.model_selection import train_test_split

indices = np.arange(len(person_ids))

train_idx, test_idx = train_test_split(
    indices,
    test_size=0.2,
    random_state=42
)

train_patients = set(person_ids[i] for i in train_idx)
test_patients = set(person_ids[i] for i in test_idx)

leaked_patients = train_patients.intersection(test_patients)

print(f"Patients appearing in BOTH train and test: {len(leaked_patients)}")
```
